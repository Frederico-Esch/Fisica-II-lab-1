```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def calcN(n):
    L = 1.79
    T = 0.49
    mi = 0.00017
    return n/(2*L) * np.sqrt(T/mi)
```

```python
calcN(1)
```

```python
df = pd.DataFrame(columns=["n", "f", "calculado"])
df
```

```python
insert = lambda x,y: {"n":x, "f":y, "calculado":calcN(x)}
```

```python
df.loc[0] = insert(1, 15)
df
```

```python
df.loc[1] = insert(2, 27)#({"n":2, "f":27})
df
```

```python
df.loc[2] = insert(3, 47)#({"n":3, "f":47})
df
```

```python
df.loc[3] = insert(4, 58) # ({"n":4, "f":58})
df
```

```python
df.loc[4] = insert(5, 82) #({"n":5, "f":82})
df
```

```python
df.to_csv("Tabela.csv", index=False)
```

```python
plt.figure(figsize=(10, 10))
plt.scatter(df["n"], df["f"])
plt.xlabel("N")
plt.ylabel("f (Hz)")
```

```python
def calcM(m):
    L = 1.79
    g = 9.8
    T = g*m
    mi = 0.00017
    n = 1
    return n/(2*L) * np.sqrt(T/mi)
```

```python
M_o = 0.05045
insert = lambda x,y: {"m": M_o+0.01*(x-1), "f":y, "calculado": calcM(M_o+0.01*(x-1))}
```

```python
calcM(M_o)
```

```python
calcM(M_o+0.01) # + 10 g
```

```python
calcM(M_o+0.02) # + 20g
```

```python
massas = pd.DataFrame(columns=["m", "f","calculado"])
```

```python
massas.loc[0] = insert(1, 15) #{"m":0.05045, "f":15}
massas
```

```python
massas.loc[1] = insert(2, 16) #{"m":M_o+0.01, "f":16}[
massas
```

```python
massas.loc[2] = insert(3, 18) #{"m":M_o+0.02, "f":18}
massas
```

```python
massas.loc[3] = insert(4, 20) #{"m":M_o+0.03, "f":20}
massas
```
